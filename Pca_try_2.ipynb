```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import sklearn
from sklearn import decomposition 
from sklearn.decomposition import PCA

%matplotlib inline
```

```python
plt.figure(figsize=(10,15))
image = np.array(Image.open('deer.jpg'))
plt.imshow(image)
plt.grid(True)
```

```python
image.shape
```

```python
def dimension_cutter(n):
    #drop 3rd dimension
    image_low = np.reshape(image, (1800,8640)) #2880 * 3 
    pca = PCA(n_components=n)
    data = pca.fit_transform(image_low)
    image_reduced = pca.inverse_transform(data)
    image_reduced = np.reshape(image_reduced, (1800,2880,3))
    print(image_reduced.shape)
    #draw picture
    plt.figure(figsize=(10,15))
    plt.imshow((image_reduced).astype(np.uint8))
    plt.title('PCA reduction n components')
    plt.grid(True)
```

```python
dimension_cutter(10)
```

```python
dimension_cutter(20)
```

```python
dimension_cutter(50)
```

```python
dimension_cutter(100)
```
