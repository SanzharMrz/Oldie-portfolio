__The same thing only in function__

```python
def dimension_cutter(n):
    im = np.array(Image.open('target_photo.jpg'))
    im2D = mean(im, 2)
    pca = PCA(n_components = n) 
    lower_dimensional_data = pca.fit_transform(im2D)
    imReduced = pca.inverse_transform(lower_dimensional_data)
    plt.figure(figsize=(20,25))
    plt.subplot(1, 2, 1)
    plt.imshow(im);
    plt.xlabel('560 x 960 components', fontsize = 14)
    plt.title('Original Image', fontsize = 20)
    grid(True)
    plt.subplot(1, 2, 2)
    plt.xlabel('10 components', fontsize = 14)
    plt.imshow(imReduced, cmap = 'binary_r')
    plt.title('After PCA', fontsize = 20)
    grid(True)
    return lower_dimensional_data.shape
```

```python
dimension_cutter(3)
```
